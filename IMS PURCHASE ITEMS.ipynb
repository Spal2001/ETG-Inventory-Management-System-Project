THE ITEMS PRESENT IN THE INVENTORY ARE SHOWN HERE

```python
import json
print("*************************************")
print("!!! WELCOME TO ALL CUSTOMERS !!!")
print("The items available are listed below: ")
print("Choose the products you want to buy")
print("*************************************")
fp=open("Record.json",'r')
txt=fp.read()
products=json.loads(txt)
fp.close()
products
```

```python
transaction=1   #initially the transaction id is set to 1
sell={transaction:{'Customer Name':'','Product Name':'','Price':0,'Quantity':0,'Amount':0.0}} #this dictionary is created to store the sells detail in this format
```

HERE THE PROGRAM FOR PURCHASE IS WRITTEN AND THE TRANSACTION ID IS INCREASED ACCORDINGLY

```python
import time
total=0  #this is the total amount initially
name=input("Enter your Name: ")
print("Name of Customer: ",name) 
n=int(input("How many do you want to buy: "))   
for i in range(0,n):  
    amt=0
    reg=input("Enter Product ID you want to buy: ")
    if(reg in products.keys()):
        rec=products[reg]
        sells=['Customer Name','Product Name','Price','Quantity','Amount','Time']
        record=[]
        print("*******************************************")
        print("Product Name: ",rec['Name'])
        u_quant=int(input("Enter quantity : "))
        if u_quant>rec['Quantity']:  #if the customer gives quantity which is more than the quantity present in the Inventory then this will execute
            print("Please enter your quantity within range ")
            continue
        print("Transaction ID :",transaction)
        print("Product Name: ",rec['Name'])
        print("Price: ",rec['Price'])
        print("Quantity:",u_quant)
        print("Quantity remaining: ",rec['Quantity']-u_quant)
        rec['Quantity']=rec['Quantity']-u_quant
        amt+=rec['Price']*u_quant
        if amt > 500:  #Here we will give discount to the user for a item if the amt for that item is greater than 500
            print("Discount Available: ",rec['Discount'])
            amt=amt-(amt*rec['Discount'])
        total+=amt #Here total amount is calculated
        print("Price: ",amt)
        print("---------------------------------------------")
        if(transaction in sell.keys()):
            res=sell[transaction]
            res['Customer Name']=name
            res['Product Name']=rec['Name']
            res['Price']=rec['Price']
            res['Quantity']=u_quant
            res['Amount']=amt
            res['Time']=time.ctime()
            transaction+=1
        else:
            record.append(name)
            record.append(rec['Name'])
            record.append(rec['Price'])
            record.append(u_quant)
            record.append(amt)
            record.append(time.ctime())
            res1=dict(zip(sells,record))
            sell[transaction]=res1
            transaction+=1
    else:
        print("Enter valid Product ID!!!!")
        print("Please start from beginning.......Program is going to Quit!")
        break
        print("**************************************************")
f=open('Sales.json','w')   #we are writing the sales made in the Sales.json file 
js=json.dumps(sell)
f.write(js)
f.close()
fd=open('Record.json','w')   #Here the Record.json file is updated again after purchasing is done with remaining quantity
r=json.dumps(products)
fd.write(r)
fd.close()
print("Your Total Amount is : ",total)  #This is the total amount to be paid
print("Printed at:",time.ctime())
print("!!! THANK YOU PLEASE VISIT AGAIN !!!")
```
