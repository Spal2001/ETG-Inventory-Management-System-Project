HERE WE CAN ADD OR UPDATE PRODUCTS INTO THE INVENTORY

```python
print("Welcome")
print("***********************************************")
n=int(input("Enter how many items yo want to add/modify: "))
for i in range(0,n):
    items=['Name','Price','Quantity','Category','Discount','Measurements','Expiry Date']
    add=[]
    reg=input("Enter product ID: ")
    if reg in products.keys():  #searches for the product id in the Inventory
        print("--------------------------------------------------")
        print("Product Exists you can only increase the quantity ")
        rec=products[reg]
        print("Product Name: ",rec['Name'])
        qn=int(input("Enter quantity of the item to be added: "))
        rec['Quantity']=rec['Quantity']+qn  #if product is present in the Inventory then we can increase the quantity
        print("New Quantity :",rec['Quantity'])
        print("Quantity Increases Successfully !!!!")
        print("-------------------------------------------------------")
    else:    #if product id is not found then it will create new product in the inventory
        print("--------------------------------------------------------")
        print("New Product to be entered")
        name=input("Enter product name :")
        add.append(name)
        pr=int(input("Enter price of the product: "))
        add.append(pr)
        qn=int(input("Enter quantity of the product :"))
        add.append(qn)
        category=input("Enter category :")
        add.append(category)
        dis=float(input("Enter discount: "))
        add.append(dis)
        measure=input("Enter measurement: ")
        add.append(measure)
        ex=input("Enter Expiry Date: ")
        add.append(ex)
        details=dict(zip(items,add))
        products[reg]=details #the new dictionary with updated items are added
        print("New item added successfully !!!!")
        print("------------------------------------------------------------")
print("!!! THANK YOU !!!")
print("The Inventory Updated Successfully ")
print("***********************************************************")
print("THE UPDATED INVENTORY PRINTED BELOW: ")
fd=open('Record.json','w')
js=json.dumps(products)  #here the updated Inventory is written again to 'Record.json' file
fd.write(js)
fd.close()
products #here the updated Inventory is printed
```
